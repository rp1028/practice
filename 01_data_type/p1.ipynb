### 문제 1. 사용자 입력 데이터 처리

어떤 설문조사 시스템에서 나이를 입력받았습니다. 입력받은 `age_input = "25"`를 이용해 10년 뒤의 나이를 계산하여 "10년 뒤 당신은 35세가 됩니다."라고 출력하는 코드를 작성하세요

```python
age_input = "25"
# 여기에 코드를 작성하세요.

# 문자열을 정수(int)로 변환한 뒤 계산
future_age = int(age_input) + 10

print(f"10년 뒤 당신은 {future_age}세가 됩니다.")
```

### 문제 2. 리스트의 중복 제거와 정렬

다음 리스트에서 중복된 과일 이름을 제거하고, 가나다 순(오름차순)으로 정렬된 리스트를 출력하세요.

```python
fruits = ["사과", "바나나", "사과", "포도", "바나나", "체리"]
# 여기에 코드를 작성하세요.

# 1. set()을 사용하여 중복 제거
unique_fruits = set(fruits)

# 2. sorted() 함수를 사용하여 가나다순(오름차순)으로 정렬 후 리스트로 반환
sorted_fruits = sorted(unique_fruits)

print(f"중복 제거 및 정렬 결과: {sorted_fruits}")

```

### 문제 3. 딕셔너리 안전하게 조회하기

아래와 같은 성적표 딕셔너리가 있습니다. 사용자가 **"과학"** 과목의 점수를 조회하려고 할 때, 에러가 발생하지 않도록 하고 데이터가 없을 경우 "해당 과목의 점수가 없습니다."라고 출력하는 코드를 작성하세요.

```python
scores = {"국어": 90, "수학": 85, "영어": 95}
# 여기에 코드를 작성하세요.

# get() 메서드를 사용하여 '과학' 점수 조회
science_score = scores.get("과학", "해당 과목의 점수가 없습니다.")

print(f"과학 점수 조회 결과: {science_score}")
```

### 문제 4. 튜플과 리스트의 결합

아래 두 튜플의 모든 요소를 하나의 리스트 `all_points`에 담아 `[10, 20, 30, 40]`이 출력되게 만드세요.

```python
point1 = (10, 20)
point2 = (30, 40)

all_points = list(point1 + point2)
print(f"결과: {all_points}")
```

사용자로부터 입력받은 검색어 `search_keyword`가 비어있는지 확인하는 코드를 작성하세요.

- 검색어가 있으면: **" '검색어'를 검색합니다."** 출력
